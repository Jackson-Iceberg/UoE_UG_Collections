```python
# Q1 a
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
%matplotlib inline
import sklearn
#print(sklearn.__version__)
data_path = os.path.join(os.getcwd(), 'data', 'regression_part1.csv')
regressionData = pd.read_csv(data_path, delimiter = ',')
regressionData.info()
```

```python
# b
from sklearn import linear_model
lrg = linear_model.LinearRegression()
revision_time = regressionData.loc[:,['revision_time']].values
exam_score = regressionData.loc[:,['exam_score']].values
lrg.fit(revision_time, exam_score)
print(lrg.intercept_)
print(lrg.coef_)
exam_predict = lrg.predict(revision_time)
#print(exam_predict)
print("残差平方和：%.10f" %np.mean((exam_predict - exam_score)**2))
```

```python
# Question c
plt.title("linear regression based on revision_time and exam_score")
plt.scatter(revision_time,exam_score,label="input data")
#plt.plot(lrg.intercept_,lrg.coef_,label="regression_line")
plt.xlabel("revision_time")
plt.ylabel("exam_score")
plt.grid(True)
plt.plot(revision_time, exam_predict,'r',label = 'linear regression')
plt.legend()
plt.savefig("Question 1c.png")
```

```python
X = np.insert(regressionData.loc[:,['revision_time']].values, 0, 1, axis=1)
X_ = np.linalg.inv(X.T.dot(X))
w = X_.dot(X.T).dot(regressionData.loc[:,['exam_score']].values)
y_pred = X.dot(w)

print(w)
print("               ")
#print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
exam_true = exam_score
exam_predict = exam_predict
mse1 = mean_squared_error(exam_true, exam_predict)
mse2 = mean_squared_error(exam_true, y_pred )
print(format(mse1,".20f"))
print(format(mse2,".20f"))
print(mse1)
print(mse2)
# 小数点在13位开始不同
```

```python
# Q2
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
%matplotlib inline
import sklearn
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import linear_model

data_path = os.path.join(os.getcwd(), 'data', 'regression_part2.csv')
nonLinearRegression = pd.read_csv(data_path, delimiter = ',')
nonLinearRegression.info()
nonLinearRegression.sort_values(by=['input'], inplace=True)
x_input = nonLinearRegression.loc[:,['input']].values
y_output = nonLinearRegression.loc[:,['output']].values


plt.figure(figsize=(10,10))
plt.xlabel("input value")
plt.ylabel("output value")
plt.title("Polynomial Nonlinear Regression Models")
plt.scatter(x_input,y_output,label="input data")
plt.grid()
for m in [1,2,3,4]:
    poly = PolynomialFeatures(m)
    x_degreed = poly.fit_transform(x_input)
    lrg = linear_model.LinearRegression(fit_intercept = False)
    lrg.fit(x_degreed,y_output)
    y_predict = lrg.predict(x_degreed)
    plt.plot(x_input,y_predict,label = "%d degree of polynomial features" % m)
plt.legend()
plt.savefig("Question 2a.png")

```

```python
# Q2 b
from sklearn.metrics import mean_squared_error
plt.figure(figsize=(12,8))
plt.title("Mean Squared Error of different degree Polynomial Nonlinear Regression Models")
plt.ylabel("Mean Squared Error Value")
plt.xlabel("Degree")
for m in [1,2,3,4]:
    poly = PolynomialFeatures(m)
    x_degreed = poly.fit_transform(x_input)
    lrg = linear_model.LinearRegression(fit_intercept = False)
    lrg.fit(x_degreed,y_output)
    y_predict = lrg.predict(x_degreed)
    mseDegree = mean_squared_error(y_output, y_predict)
    plt.bar(m,mseDegree,width=0.8, bottom=None,label ="%d degree Mean Squared Error" %m)
    print(mseDegree)
    mseDisplay = format(mseDegree,".5f")
    plt.text(m,mseDegree,"MSE value= "+mseDisplay,horizontalalignment='center')
plt.legend()
plt.savefig("Question 2b")
```

```python
# Q2 d
import math

plt.figure(figsize=(10,6))
plt.scatter(x_input,y_output,label="training points")
plt.title('Radial Basis Function Model(RBF) in different alpha')
plt.xlabel('input value')
plt.ylabel('output value')
plt.grid()
for a in [0.2,100,1000]:
    x_rbf = []
    for x in x_input:
        x_mid = [1]
        for c in [-4.0,-2.0, 2.0, 4.0]:
            rbf_value = (math.exp(-0.5*(pow((x-c),2)/pow(a,2))))
            x_mid.append(rbf_value)
        x_rbf.append(x_mid)
    lrg = linear_model.LinearRegression(fit_intercept = False)  
    lrg.fit(x_rbf,y_output)
    y_predict = lrg.predict(x_rbf)        
    alpha = ("%.1f" %a)
    plt.plot(x_input,y_predict,label="radial basis functions(RBF) when Alpha = " + alpha)       
plt.legend()     
plt.savefig('Question 2d.png')
```

```python
# Q4
import numpy as np
from sklearn.metrics import accuracy_score

data_path = os.path.join(os.getcwd(), 'data', 'classification_eval_1.csv')
classificationData = pd.read_csv(data_path, delimiter = ',')
classificationData.info()
gt = classificationData.pop("gt")
alg_1 = classificationData.pop("alg_1")
alg_2 = classificationData.pop("alg_2")
alg_3 = classificationData.pop("alg_3")
alg_4 = classificationData.pop("alg_4")

alg_1 = [1 if alg_1[i]>=0.5 else 0 for i in range(len(alg_1))]
alg_2 = [1 if alg_2[i]>=0.5 else 0 for i in range(len(alg_2))]
alg_3 = [1 if alg_3[i]>=0.5 else 0 for i in range(len(alg_3))]
alg_4 = [1 if alg_4[i]>=0.5 else 0 for i in range(len(alg_4))]

alg1_accuracy = accuracy_score(gt,alg_1)
alg2_accuracy = accuracy_score(gt,alg_2)
alg3_accuracy = accuracy_score(gt,alg_3)
alg4_accuracy = accuracy_score(gt,alg_4)
print(alg1_accuracy)
print(alg2_accuracy)
print(alg3_accuracy)
print(alg4_accuracy)

# incerase 0.5 value 可以同时增加他们的accuracy
```

```python
# Q4 c
import numpy as np
from sklearn.metrics import roc_curve


data_path = os.path.join(os.getcwd(), 'data', 'classification_eval_1.csv')
classificationData = pd.read_csv(data_path, delimiter = ',')
classificationData.info()
gt = classificationData.pop("gt")
alg_1 = classificationData.pop("alg_1")
alg_2 = classificationData.pop("alg_2")
alg_3 = classificationData.pop("alg_3")
alg_4 = classificationData.pop("alg_4")

print(alg_3)
fpr_1, tpr_1, thresholds_1 = roc_curve(gt,alg_1)
fpr_2, tpr_2, thresholds_2 = roc_curve(gt,alg_2)
fpr_3, tpr_3, thresholds_3 = roc_curve(gt,alg_3)
fpr_4, tpr_4, thresholds_4 = roc_curve(gt,alg_4)



plt.plot(fpr_1,tpr_1,label="alg_1(area = 0.732)")
plt.plot(fpr_2,tpr_2,label="alg_2(area = 0.632)")
plt.plot(fpr_3,tpr_3,label="alg_3(area = 0.064)")
plt.plot(fpr_4,tpr_4,label="alg_4(area = 0.847")
plt.title(" ROC curves for four models")
plt.xlabel("False Positive Rate")
plt.ylabel("True positive Rate")
plt.legend()
plt.savefig("Question 4c.jpg")
plt.show()
```

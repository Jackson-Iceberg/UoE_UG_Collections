```python
# 1.2

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal','Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# arrayCollection used to divide each data into typical group
arrayCollection = [[],[],[],[],[],[],[],[],[],[]]

for i in range(60000):
    if (Ytrn[i] == 0):
        arrayCollection[0].append(Xtrn[i])
    if (Ytrn[i] == 1):
        arrayCollection[1].append(Xtrn[i])
    if (Ytrn[i] == 2):
        arrayCollection[2].append(Xtrn[i])
    if (Ytrn[i] == 3):
        arrayCollection[3].append(Xtrn[i])
    if (Ytrn[i] == 4):
        arrayCollection[4].append(Xtrn[i])
    if (Ytrn[i] == 5):
        arrayCollection[5].append(Xtrn[i])
    if (Ytrn[i] == 6):
        arrayCollection[6].append(Xtrn[i])
    if (Ytrn[i] == 7):
        arrayCollection[7].append(Xtrn[i])
    if (Ytrn[i] == 8):
        arrayCollection[8].append(Xtrn[i])
    if (Ytrn[i] == 9):
        arrayCollection[9].append(Xtrn[i])

# len(arrayCollection[0])  check the size of arrayCollection      

# This array used to store the index of two closest samples and two furthest samples of that class
IndexCollection = [[],[],[],[],[],[],[],[],[],[]]

for i in range(10):
    array_mean = np.mean(arrayCollection[i],axis=0)
    Mini_Index = 0
    Mini2_Index = 0
    Max_Index = 0
    Max2_Index = 0
    
    # get mini_index
    array1_dist = np.linalg.norm(arrayCollection[i][0] - array_mean)
    for j in range(6000):
        if (np.linalg.norm(arrayCollection[i][j] - array_mean) < array1_dist):
            array1_dist = np.linalg.norm(arrayCollection[i][j] - array_mean)
            Mini_Index = j
#     print(array1_dist)
#     print(Mini_Index)
    
    array2_dist = np.linalg.norm(arrayCollection[i][0] - array_mean)
    for j in range(6000):
        if (j != Mini_Index):
            if (np.linalg.norm(arrayCollection[i][j] - array_mean) < array2_dist):
                array2_dist = np.linalg.norm(arrayCollection[i][j]- array_mean)
                Mini2_Index = j
#     print(array2_dist)
#     print(Mini2_Index)
    
    array3_dist = np.linalg.norm(arrayCollection[i][0] - array_mean)                                        
    for j in range(6000):
        if (np.linalg.norm(arrayCollection[i][j] - array_mean) > array3_dist):
            array3_dist = np.linalg.norm(arrayCollection[i][j]- array_mean)
            Max_Index = j
#     print(array3_dist)
#     print(Max_Index)
    
    array4_dist = np.linalg.norm(arrayCollection[i][0] - array_mean)
    for j in range(6000):
        if (j != Max_Index):
            if (np.linalg.norm(arrayCollection[i][j] - array_mean) > array4_dist):
                array4_dist = np.linalg.norm(arrayCollection[i][j]- array_mean)
                Max2_Index = j
    # print(array4_dist) 
    # print(Max2_Index)
    
    
    
    # index 0,1 is two closest samples
    # index 2,3 is two furthest samples
    
    IndexCollection[i].append(Mini_Index)
    IndexCollection[i].append(Mini2_Index)
    IndexCollection[i].append(Max_Index)
    IndexCollection[i].append(Max2_Index)  
    
    
    
```
